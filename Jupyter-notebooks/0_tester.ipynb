```python
def clean_categorical_values(df):
    """
    Standardize categorical values and fix typos
    """
    df_clean = df.copy()
    
    # Define standardization mappings for each categorical variable
    standardization_maps = {
        'livestock_owned': {
            'yes': 'Yes', 'y': 'Yes', 'YES': 'Yes', 'Y': 'Yes',
            'no': 'No', 'n': 'No', 'NO': 'No', 'N': 'No'
        },
        'usedOM': {
            'yes': 'Yes', 'y': 'Yes', 'YES': 'Yes', 'Y': 'Yes',
            'no': 'No', 'n': 'No', 'NO': 'No', 'N': 'No'
        },
        'climate_info_received': {
            'yes': 'Yes', 'y': 'Yes', 'YES': 'Yes', 'Y': 'Yes',
            'no': 'No', 'n': 'No', 'NO': 'No', 'N': 'No'
        },
        'maize_unit': {
            'kg': 'kg', 'KG': 'kg', 'kgs': 'kg', 'KGS': 'kg',
            'tons': 'tons', 'Tons': 'tons', 'TONS': 'tons',
            'bags': 'bags', 'Bags': 'bags', 'BAGS': 'bags'
        }
    }
    
    # Apply standardization
    for column, mapping in standardization_maps.items():
        if column in df_clean.columns:
            df_clean[column] = df_clean[column].replace(mapping)
            # Also handle case sensitivity
            df_clean[column] = df_clean[column].str.title()  # Or use specific mapping
    
    return df_clean

# Apply categorical cleaning
merged_clean = clean_categorical_values(merged)
```
